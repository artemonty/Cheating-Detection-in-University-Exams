```python
user_to_user_data_exam_1 = user_answer_exam_1.merge(user_answer_exam_1, on='task_ind', suffixes=('_1', '_2'))
```

```python
user_to_user_data_exam_1['key'] = user_to_user_data_exam_1.apply(lambda x: (int(min(x['user_id_1'], x['user_id_2'])), int(max(x['user_id_1'], x['user_id_2']))), axis=1)
```

```python
user_to_user_data_exam_1 = user_to_user_data_exam_1[user_to_user_data_exam_1['user_id_1'] != user_to_user_data_exam_1['user_id_2']]
```

```python
user_similarity_frac = user_to_user_data_exam_1.groupby(['user_id_1', 'user_id_2'], as_index=False) \
    .agg({'verdict_1': lambda x: len(x) / 16}) \
    .rename(columns={'verdict_1': 'similarity'})
```

```python
user_similarity_frac
```

```python
sns.histplot(user_similarity_frac['similarity'], bins=16)
```

```python
user_similarity_frac['similarity'].value_counts().sort_index()
```

```python
user_similarity_frac.groupby('user_id_1', as_index=False).agg({'similarity': 'max'})['similarity'].value_counts().sort_index()
```

```python
sns.histplot(user_similarity_frac.groupby('user_id_1', as_index=False).agg({'similarity': 'max'})['similarity'], binwidth=0.01)
```

```python
suspicious_users_exam_1 = user_similarity_frac[user_similarity_frac['similarity'] >= 0.85].copy(deep=True)
```

```python
suspicious_users_exam_1
```

```python
suspicious_users_exam_1.user_id_1.nunique()
```

```python
user_success_df = user_answer_exam_1[user_answer_exam_1['user_id'].isin(
    suspicious_users_exam_1['user_id_1'])] \
    .groupby('user_id', as_index=False) \
    .agg({'verdict': lambda x: (x == "OK") \
    .sum() / len(x)}).rename(columns={'verdict': 'success_rate'})
```
